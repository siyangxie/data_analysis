# 1 tag_A_test_20211230154013.log

e.g.

````python
name = "tag_A_test_20211230154013.log"
fr_path = folder / name
fw_path = folder / "fixed" / "tag_A_test_20211230154013.txt"

def preprocess(file_read, file_write, debug=False):
  with open(file_read, "r") as fr, open(file_write, "w") as fw:
    lines = fr.readlines()
    x = []  # 时间
    y = []  # 定位

    for i, c in enumerate(lines):
      if c.startswith("["):
        x.append(c[1:24])

      if c[0].isdigit():
        c = c.split(" ")
        if c[4] != "05" or c[2:4] != ["0A", "00"]:  # 命令字段
          x.pop()
          continue
        
        if debug:
          print("raw data: ", c)
        pos = c[8:12]  # 人员定位标签位置信息报文(TLU_PLT_TWR2CCU)
        if debug:
          print("raw loc data:", pos)

        pos = "".join(pos)
        pos = bytearray.fromhex(pos)
        if debug:
          print("hex:", pos)

        # pos.reverse()
        # print("reversed:", pos)

        pos = struct.unpack("<I", pos)[0]
        if debug:
          print("location (dm):", pos)
        y.append(str(pos))

        if debug:
          break

        fw.write(x[-1])
        fw.write(",")
        fw.write(y[-1])
        fw.write("\n")

    print("x: ", len(x))
    print("y: ", len(y))



preprocess(file_read=fr_path, file_write=fw_path, debug=True)
````

````python
if os.path.exists(fw_path):
  os.remove(fw_path)
preprocess(file_read=fr_path, file_write=fw_path)
````

# 3 tag_AandB_test20211230182209.log

````python
name = "tag_AandB_test20211230182209.log"
fr_path = folder / name
fw_path = folder / "fixed" / "tag_AandB_test20211230182209.txt"

def preprocess(file_read, file_write, debug=False):
  with open(file_read, "r") as fr, open(file_write, "w") as fw:
    lines = fr.readlines()
    x = []  # 时间
    y = []  # 定位
    tag = ""

    for i, c in enumerate(lines):
      if c.startswith("["):
        x.append(c[1:24])

      if c[0].isdigit():
        c = c.split(" ")
        if c[4] != "05":  # 命令字段
          x.pop()
          continue

        if c[2:4] == ["0B", "00"]:
          tag = "B"
        elif c[2:4] == ["0A", "00"]:
          tag = "A"
        else:
          continue
        
        if debug:
          print("tag :", tag)
          print("raw data: ", c)
        pos = c[8:12]  # 人员定位标签位置信息报文(TLU_PLT_TWR2CCU)
        if debug:
          print("raw loc data:", pos)

        pos = "".join(pos)
        pos = bytearray.fromhex(pos)
        if debug:
          print("hex:", pos)

        # pos.reverse()
        # print("reversed:", pos)

        pos = struct.unpack("<I", pos)[0]
        if debug:
          print("location (dm):", pos)
        y.append(str(pos))

        if debug:
          break

        fw.write(x[-1])
        fw.write(",")
        fw.write(tag)
        fw.write(",")
        fw.write(y[-1])
        fw.write("\n")

    print("x: ", len(x))
    print("y: ", len(y))



preprocess(file_read=fr_path, file_write=fw_path, debug=True)
````

````python
if os.path.exists(fw_path):
  os.remove(fw_path)
preprocess(file_read=fr_path, file_write=fw_path)
````
